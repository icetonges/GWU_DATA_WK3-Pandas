```python
# Dependencies
import pandas as pd
```

```python
# load CSV
newcoders_csv = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
newcoders_pd = pd.read_csv(newcoders_csv, encoding = 'ISO-8859-1', low_memory=False)
newcoders_pd.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_newcoders_pd = newcoders_pd.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_newcoders_pd.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
reduced_newcoders_pd = reduced_newcoders_pd.replace({0.0:"No", 1.0:"Yes"})
reduced_newcoders_pd.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_pd = reduced_newcoders_pd.loc[reduced_newcoders_pd["AttendedBootcamp"] == "Yes", :]
attended_pd.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
#Bootcampattend = attended_pd["BootcampName"].value_counts()
#number_attended_by_bootcamp = pd.DataFrame(number_attended_by_bootcamp.groupby("BootcampName"))
#number_attended_by_bootcamp.columns=["BootcampName","Bootcampattend"]
#number_attended_by_bootcamp.reset_index(inplace=True)
#number_attended_by_bootcamp
#number_attended_by_bootcamp = pd.DataFrame({"Bootcampattend":[Bootcampattend],"BootcampName":[BootcampName]})
#type(Bootcampattend)
number_attended_by_bootcamp = pd.DataFrame(attended_pd["BootcampName"].value_counts())
number_attended_by_bootcamp.reset_index(inplace=True)
reduced_number_attended_by_bootcamp = number_attended_by_bootcamp.loc[(number_attended_by_bootcamp["BootcampName"] > 1)]
reduced_number_attended_by_bootcamp = reduced_number_attended_by_bootcamp.rename(
    columns={"index": "Bootcamp Name","BootcampName":"Number Attends" })
reduced_number_attended_by_bootcamp.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommended_pd = attended_pd.loc[attended_pd["BootcampRecommend"] == "Yes", :]
recommended_pd = pd.DataFrame(recommended_pd["BootcampName"].value_counts())
recommended_pd.reset_index(inplace=True)
recommended_pd = recommended_pd.rename(columns={"index": "Bootcamp Name","BootcampName":"Number Recommended" })
recommended_pd = recommended_pd.loc[(recommended_pd["Number Recommended"] > 1)]
recommended_pd.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_pd = reduced_number_attended_by_bootcamp.merge(recommended_pd, on="Bootcamp Name")
merge_pd.head()
#merge_pd.dtypes
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
#number_attended = merge_pd["Number Attends"]
#number_recommended = merge_pd["Number Recommended"]
#recommended_percent = number_recommended/number_attended *100
#recommended_percent
merge_pd['Recommended_percent'] = merge_pd["Number Recommended"]/merge_pd["Number Attends"]*100
sorted_pd=merge_pd.sort_values("Recommended_percent",ascending=False)
sorted_pd["Recommended_percent"] =sorted_pd["Recommended_percent"].map("{0:,.2f}%".format)
sorted_pd.head()
```

```python
# Export to excel and remove index
sorted_pd.to_excel("output/sorted_pd.xlsx", index=False)
```
